```python
import numpy as np
import matplotlib.pyplot as plt

rho = 7
sigma_min = 0.002
sigma_max = 80

# First function
def karras_boundaries_v1(num_timesteps):
    rho_inv = 1.0 / rho 
    steps = np.arange(num_timesteps) / max(num_timesteps - 1, 1)
    sigmas = sigma_min**rho_inv + steps * (
            sigma_max**rho_inv - sigma_min**rho_inv
        )
    sigmas = sigmas**rho
    return sigmas

# Second function
def karras_boundaries_v2(num_timesteps):
    ramp = np.linspace(1, 0, num_timesteps)
    min_inv_rho = sigma_min ** (1 / rho)
    max_inv_rho = sigma_max ** (1 /rho)
    sigmas = (max_inv_rho + ramp * (min_inv_rho - max_inv_rho)) ** rho 
    return sigmas 

# Test both functions
num_timesteps = 1000
sigmas_v1 = karras_boundaries_v1(num_timesteps)
sigmas_v2 = karras_boundaries_v2(num_timesteps)

# Plot the results
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(sigmas_v1, label='karras_boundaries_v1')
plt.title('karras_boundaries_v1')
plt.xlabel('Timesteps')
plt.ylabel('Sigma')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(sigmas_v2, label='karras_boundaries_v2')
plt.title('karras_boundaries_v2')
plt.xlabel('Timesteps')
plt.ylabel('Sigma')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
